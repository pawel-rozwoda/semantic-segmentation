```python
import pandas as pd
import matplotlib.pyplot as plt

train_losses = pd.read_csv('model_01|23|2020--13:17:59/train_losses.csv', index_col=False, header = None)
train_losses.columns = ['epoch','total','dice','bce']

test_losses = pd.read_csv('model_01|23|2020--13:17:59/test_losses.csv', index_col=False, header = None)
test_losses.columns = ['epoch','total','dice','bce']

```

# Test vs Train

```python
plt.plot( 'epoch', 'total', data=train_losses, marker='', color='olive', linewidth=2, linestyle='dashed', label="train")
plt.plot( 'epoch', 'total', data=test_losses, marker='', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="test")
plt.legend()
```

# Train losses

```python
plt.plot( 'epoch', 'dice', data=train_losses, marker='', color='olive', linewidth=2, linestyle='dashed', label="train dice")
plt.plot( 'epoch', 'bce', data=train_losses, marker='', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="train bce")
plt.legend()
```

# Test losses

```python
plt.plot( 'epoch', 'dice', data=test_losses, marker='', color='olive', linewidth=2, linestyle='dashed', label="test dice")
plt.plot( 'epoch', 'bce', data=test_losses, marker='', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="test bce")
plt.legend()
```
